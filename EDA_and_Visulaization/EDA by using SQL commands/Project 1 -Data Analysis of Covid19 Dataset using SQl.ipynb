### 1) Show the number of Confirmed, Deaths and Recovered cases in each Region.

```python
%sql select Region,Confirmed,Deaths,Recovered from covid_data group by Region limit 10 # 'group by' command group the data in resectice distinct sets
```

### 2) Remove all the records where the Confirmed Cases is Less Than 10.

```python
%sql delete from covid_data where Confirmed<10 # We use 'delete' command to delete particular row satisfying condition .#To delete the columns from the table we use 'drop' command'
```

```python
%sql select count(*) from covid_data
```

#### Here we can see the total no. of rows have reduced from 321 to 304 as we perform the delete operation. We can verify our given commands by this.

### 3) In which Region, maximum number of Confirmed cases were recorded ?

```python
%sql select region,confirmed from covid_data order by confirmed desc limit 1
```

### 4) In which Region, minimum number of Deaths cases were recorded ?

```python
%sql select region,deaths from covid_data order by deaths asc limit 50
```

### 5) How many Confirmed, Deaths & Recovered cases were reported from India till 29 April 2020 ?

### 6) Sort the entire data wrt No. of Confirmed cases in ascending order.

```python
%sql select * from covid_data order by confirmed asc limit 10  # 'order by' clause to arrange data in ascending or descending order 
```

### 7) Sort the entire data wrt No. of Recovered cases in descending order.

```python
%sql select * from covid_data order by Recovered desc limit 10
```
