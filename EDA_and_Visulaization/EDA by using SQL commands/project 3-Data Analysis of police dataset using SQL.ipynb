### 1) Find no. of cases by race 

```python
%sql select distinct(driver_race),count(driver_race) from police_data group by driver_race
```

### 2) For Speeding , were Men or Women stopped more often ? 

```python
%sql select driver_gender,count(driver_gender) from police_data where violation like '%speeding%' group by driver_gender
```

### 3) Does gender affect who gets searched during a stop ?

```python
%sql select distinct(driver_gender),count(search_conducted) from police_data group by driver_gender 
```

### 4) What is the mean stop_duration ?

```python
%sql select avg(stop_duration) as mean from police_data 
```

### another method

```python
%sql select distinct stop_duration ,avg(stop_duration) as mean from police_data group by stop_duration
```

### 5) Compare the age distributions for each violation.

```python
%sql show columns from police_data
```

```python
%sql select distinct(violation),avg(driver_age) from police_data group by violation
```

### 6) find distinct violation raw 

```python
%sql select distinct(violation_raw) from police_data order by violation_raw
```

### 7) Find male,female drivers having age below 20

```python
%sql select * from police_data where driver_age <20 order by driver_gender limit 10
```

### 8)Find the driveres were arrested for drug related cases

```python
%sql select * from police_data where (is_arrested != 'FALSE' and  drugs_related_stop !='FALSE')
```

### 9) Find cases of black people being searched on mere suspension 

```python
%sql select * from police_data where (driver_race='black' and search_type like '%Suspicion%') limit 10
```
