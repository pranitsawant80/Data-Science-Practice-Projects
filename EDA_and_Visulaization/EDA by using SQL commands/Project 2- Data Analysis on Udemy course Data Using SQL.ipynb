### 1) What are all different subjects for which Udemy is offering courses ?

```python
%sql select distinct subject from udemy_courses
```

### 2) Which subject has the maximum number of courses.

```python
%sql select distinct(subject),count(subject) from udemy_courses group by subject;
```

### 3) Show all the courses which are Free of Cost.

```python
%sql select * from udemy_courses where is_paid !='TRUE' limit 10
```

### 4) Show all the courses which are Paid.

```python
%sql select * from udemy_courses where is_paid='TRUE' limit 10
```

### 5) Which are Top Selling Courses ?

```python
%sql select * from udemy_courses order by num_subscribers desc limit 10
```

### 6) Which are Least Selling Courses ?

```python
%sql select * from udemy_courses order by num_subscribers asc limit 10
```

### 7) Show all courses of Graphic Design where the price is below 100 ?

```python
%sql select * from udemy_courses where subject='Graphic Design' and price < '100'
```

### 8) List out all the courses that are related to 'Python'.

```python
%sql select * from udemy_courses where course_title like '%Python%' limit 10
```

### 9) What is the Max. Number of Subscribers for Each Level of courses ?

```python
%sql select level,num_subscribers from udemy_courses group by level
```
