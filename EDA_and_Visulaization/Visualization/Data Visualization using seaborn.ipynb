## 1) Draw countplot of gender column
### countplot help us to count no. of different types of values in the column

```python
df.gender.value_counts()
```

```python
sns.countplot(df.gender)
plt.show() 
```

## Conclusion:- We can see the no. of female students are slightly more than male students.

## 2) Scatter plot of Maths score 
### scatter plot help us to understand data clusturs which could be helpful to arrive at particular conclusion

```python
sns.set(rc={"figure.figsize":(15, 8)}) # To set figure size in seaborn

# We can show titles ,lables either the way below or using matplotlib library

ax=sns.scatterplot(data=df['math score'],color='black',marker="+")
ax.set_xlabel('No of students', fontsize=20) 
ax.set_ylabel('Marks in maths paper', fontsize= 20)
ax.set_title('No of students vs Marks in maths paper',fontsize= 26) 

plt.show()
```

## Conclusion:- We can observe that the scarter plot is more dense between the marks 40 to 90 which means most of the students get marks above 40 and below 90

## 3) Plot boxplot of reading score
### Box Plot helps to find out outliers in the dataset

```python
sns.boxplot(data=df['reading score'],color='orange')
# we can make combine use of matplotlib to show labels and titles in the figure
plt.title('Boxplot of reading score', fontsize= 26)  
plt.ylabel('Reading Score',fontsize=20)
plt.show()
```

## Conclusion:-  Nearly all the values of reading score lie between 60 to 80 marks. We have some student with 100 marks and also the student with below 20 marks

## 4) Plot distplot of maths score
### distribution plot helps us to knon continuous distribution of objects

```python
sns.distplot(x = df['math score'], bins = 10,color='red')
plt.title('Distribution plot of maths score', fontsize= 20)  
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('Density',fontsize=20)
plt.show()
```

## Conclusion:- The graph shows , students got marks between 20 to 100. Also the students with marks 60-70 are the highest in number.

## 5) Plot Barplot of Parents Education Level
### Barplot helps us to easily understand the trends and no. in data

```python
Parents_education=df['parental level of education'].value_counts()
x=Parents_education.keys()
```

```python
sns.set(rc={"figure.figsize":(20, 10)}) 
ax=sns.barplot(x,Parents_education)

ax.set_xticklabels(x,rotation = 30, fontsize= 20)
ax.set_xlabel('Parents Eduactaion', fontsize=20) 

ax.set_ylabel('No of parents having the eduacation', fontsize= 20)
ax.set_title('Parents Eduactaion vs No of Parents',fontsize= 26)

plt.show()
```

## Conclusion:- The Bar graph shows most of the parents are educated from some college with no specific degree. It is followed by Associate's Degree. The parents with master's degree are least in no.

## 6) Plot Stripplot of Parents level of education and Writing Score
### Stripplot helps us to present data in clusters . which simplifies complex data and gives meaningful insights.

```python
ax=sns.stripplot(x=df['parental level of education'],y=df['writing score'])

ax.set_xticklabels(x,rotation = 30, fontsize= 20)
ax.set_xlabel('Parents level of Eduactaion', fontsize=20) 

ax.set_ylabel('writing score', fontsize= 20)
ax.set_title('Parents Eduactaion vs writing score',fontsize= 26)
plt.show()
```

## Conclusion:- Very few students (with their parent's educational qualification of bachelor's and Master's degree)  were able to get marks between 80 to 90 .
