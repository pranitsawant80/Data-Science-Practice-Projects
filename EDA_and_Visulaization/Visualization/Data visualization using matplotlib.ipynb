## 2) Bar plot of top 10 players who were got player of the match award
### Which would help us to find out the best player in the tournament . It helps us to select player in the IPL auction.

## 3) Pie Chart of Tems won Total IPL matches from 2008 to 2020 by taking toss decision 'bat first'.
### It would show us which team utilize toss winning opportunity ,choose bat first and won the match. We can conclude which team has highest ratio of winning match by bat first.

```python
win_by_bat_first=df['winner'].groupby(df['toss_decision']=='bat').value_counts().keys() 
y=df['winner'].groupby(df['toss_decision']=='bat').value_counts()
```

```python
#making pie chart 
plt.figure(figsize=(15,15))
plt.pie(y,labels=win_by_bat_first,autopct='%0.1f%%')
plt.title("IPL winners from 2008 to 2020 by bat first",fontsize=30)
plt.show()
```

### Conclusion:- Mumbai Indians, Kolkata Knight Riders and Royal Challengers Banglore are the top three performers to win toss and also win by bat first. where as Gujarat Lions ,Pune Worriors,Delhi Capitals have lower rate to win the matches.

### 5) To draw bar graph to show how many time a  particular team won the toss and also won the match.
#### We will check wether the toss win has any role in match win.

```python
df1=df[['toss_winner','winner']].sort_values('toss_winner')
toss_win=df1.toss_winner.value_counts(sort = False).keys()
print(toss_win)

df2=df[['toss_winner','winner']].sort_values('winner')
win=df2.winner.value_counts(sort=False).keys()
print(win)
```

```python
col1=df1.toss_winner.value_counts(sort = False)
col2=df2.winner.value_counts(sort=False)

x = np.arange(len(toss_win))  # the label locations
width = 0.35  # the width of the bars
fig, ax = plt.subplots(figsize=(20,10))
bar1 = ax.bar(x - width/2,col1, width, label='toss_winner')
bar2 = ax.bar(x + width/2, col2, width, label='match win')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('No of times toss win and match win',fontsize=(20))
ax.set_xlabel('Team Name',fontsize=(20))
ax.set_title('Team Name vs No of times toss win and match win',fontsize=(26))
ax.set_xticks(x, win,rotation=80,fontsize=20)

ax.legend()

ax.bar_label(bar1, padding=3,fontsize=20)
ax.bar_label(bar2, padding=3,fontsize=20)
plt.legend(['toss win','match win'],fontsize=20)
plt.show()
```

## Conclusion:- From the graph we can conclude that, toss win has significant impact on match win, as the in the graph the bars are of nearly same height of the each other withe slight differences.

## 6) To draw stacked graph of  No. of times team played matches and how many times it's won or loose

```python
Total_matches= (df.team1.value_counts())+(df.team2.value_counts())
print('Total_matches')
print(Total_matches)
print('_______________________________________________________________')
Match_win = df2.winner.value_counts(sort=False)
Match_lost = (Total_matches - Match_win)
print('Match_lost')
print(Match_lost)
```

```python
x = Total_matches.keys()
y = Total_matches
y1 = Match_lost
x2= y-y1
```

```python
plt.figure(figsize=(30,15))
plt.barh(x,y,color='green')
for index, value in enumerate(y):
    plt.text(value, index,str(value),fontsize=30)
plt.barh(x,y1,color='brown')
for index, value in enumerate(y1):
    plt.text(value, index,str(value),fontsize=30)
    
plt.barh(x,x2,color='red')
for index, value in enumerate(x2):
    plt.text(value, index,str(value),fontsize=30)
    
plt.xlabel('No. Of Matches',fontsize=30)
plt.ylabel('Name of Teams', fontsize=30)
plt.yticks(fontsize=30)
plt.title('Matches win or loose out of total matches graph', fontsize=36)
plt.legend(['match win','tie/undecide','match lost'],fontsize=20)

plt.show()
```
