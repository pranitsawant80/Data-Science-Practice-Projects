#### Although graphs take long time to code but if the data is complex (not this one) then see how graphs will make it easier to find answer of required question quickly

```python
y=df['Wind Speed_km/h'].value_counts()
ax=y.plot(kind='barh',color='green',figsize=(20, 10))
ax.set_xlabel('No. of Times')
ax.set_ylabel('Wind Speed in km/hr')
ax.set_title('Wind Speed vs No. of Times barh graph')
for index, value in enumerate(y):
    ax.text(value, index,str(value))
```

## Just to know how visibility affected in each weather condition 
### we actually used calculated-mean of visibility in each weather condition and then plot a bar graph.

### such visualization of graph to find relationships helps to build machine learning model as it clearly indicated  precise independent variables for modeling.

```python
y=df.groupby('Weather Condition').mean()
V_mean=y.Visibility_km
ax=V_mean.plot(kind='bar',color='blue',figsize=(20, 10))

ax.set_xlabel('Weather Condition')
ax.set_ylabel('Visibility in Km')
ax.set_title('Weather Condition vs Visibility in Km')
```

#### Answer= Highest visibility in mainly clear weather and Lowest visibility is in Moderate snow,blowing snow
