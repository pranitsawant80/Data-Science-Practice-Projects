```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url='https://www.wonderslist.com/10-most-powerful-militaries/'
```

```python
page=requests.get(url)
page
```

```python
content=page.content
content
```

```python
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify()) 
```

```python
soup.find_all('title')
```

```python
classes_tag=soup.find('div',class_='entry-content')
classes_tag
```

### Here we have created two dataframes to adjust with sourse code .We will then join both the table using merge (by indices)

```python
h=classes_tag.find_all('h2')
Military_Data1 =pd.DataFrame(columns=['Military_Rank','Country_Name'])
h.reverse()                    # To reverse the list as the website contains list in ascending order
for i in h:
    Country_Name=(i.get_text(strip=True).split('.')[1])  # We use split to separate data in string separated by '.
    Country_Rank=(i.get_text(strip=True).split('.')[0])
    Military_Data1= Military_Data1.append({'Military_Rank':Country_Rank,'Country_Name':Country_Name},ignore_index=True)

Military_Data1
```

```python
Military_Data2 = pd.DataFrame(columns=['Power_Index_rating','Total_population','Total_military_personnel','Aircraft','Fighter_Aircraft','Dedicated_Attack','Attack_Helicopters','Combat_tanks','Total_naval_assets','Defense_budget'])

ul=classes_tag.find_all('ul')
del ul[0:2]            # Here we delete the data at indices 0 and 1 to avoid errors by making adjustment in the list we get in above command
ul.reverse()

for u in ul:
    data = u.find_all("li")
    
    if (data !=[]):
        
        Power_Index_rating=(data[0].get_text(strip=True).split(':')[1]) 
        Total_population=(data[1].get_text(strip=True).split(':')[1]) 
        Total_military_personnel=(data[2].get_text(strip=True).split(':')[1]) 
        Aircraft=(data[3].get_text(strip=True).split(':')[1]) 
        Fighter_Aircraft=(data[4].get_text(strip=True).split(':')[1]) 
        Dedicated_Attack=(data[5].get_text(strip=True).split(':')[1]) 
        Attack_Helicopters=(data[6].get_text(strip=True).split(':')[1]) 
        Combat_tanks=(data[7].get_text(strip=True).split(':')[1]) 
        Total_naval_assets=(data[8].get_text(strip=True).split(':')[1]) 
        Defense_budget=(data[9].get_text(strip=True).split(':')[1])
        Military_Data2 = Military_Data2.append({'Power_Index_rating':Power_Index_rating,'Total_population':Total_population,'Total_military_personnel':Total_military_personnel,'Aircraft':Aircraft,'Fighter_Aircraft':Fighter_Aircraft,'Dedicated_Attack':Dedicated_Attack,'Attack_Helicopters':Attack_Helicopters,'Combat_tanks':Combat_tanks,'Total_naval_assets':Total_naval_assets,'Defense_budget':Defense_budget},ignore_index=True)
       
Military_Data2
```

```python
Military_Data =pd.merge(Military_Data1,Military_Data2, left_index=True,right_index=True)  # We use merge command to join the two dataframes
Military_Data
```

```python
# To save dataset into excel
Military_Data.to_csv('World Military Ranking')
```
